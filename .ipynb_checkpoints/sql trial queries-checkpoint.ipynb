```python
r=c.execute('''


SELECT author, group_concat(body), count(body) AS num_posts FROM posts p

WHERE EXISTS(
SELECT 1 FROM posts WHERE author=p.author AND subreddit='AskReddit')

GROUP BY author HAVING num_posts>=100

LIMIT 10;''')
r=r.fetchone()
```
